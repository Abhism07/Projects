```python
def normalize(feature):
    feature =feature - np.mean(feature)/np.std(feature)
    return feature
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
```

```python
y= data['Price']
```

```python
x=data.drop(['Price'],axis='columns')
```

```python
year = np.array(x['Year'])
year=year.reshape(-1,1)
```

```python
x['Year'] =scaler.fit_transform(year)
x['Year'] 
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
print("The first five predicted values are:")
print(y_pred[0:5])

print("The first five test values are:")
print(y_test[0:5])
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,x,y,cv=10))
print(np.average(cross_val_score(lr,x,y,cv=10)))
```

```python
from sklearn.linear_model import Lasso
lasso=Lasso()
lasso.fit(x_train,y_train)
y_pred = lasso.predict(x_test)
print("The first five predicted values are:")
print(y_pred[0:5])

print("The first five test values are:")
print(y_test[0:5])
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(lr,x,y,cv=10))
print(np.average(cross_val_score(lr,x,y,cv=10)))
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train,y_train)
y_pred = ridge.predict(x_test)
print("The first five predicted values are:")
print(y_pred[0:5])

print("The first five test values are:")
print(y_test[0:5])
```

```python
r2=r2_score(y_test,y_pred)
r2
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
y_pred = dtr.predict(x_test)
print("The first five predicted values are:")
print(y_pred[0:5])

print("The first five test values are:")
print(y_test[0:5])
```

```python
r2=r2_score(y_test,y_pred)
r2
```
